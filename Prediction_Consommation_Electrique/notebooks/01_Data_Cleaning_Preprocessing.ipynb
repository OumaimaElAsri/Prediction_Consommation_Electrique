# 01 - Data Cleaning & Preprocessing

Ce notebook charge les données météo et RTE, exécute le pipeline de nettoyage/fusion,
et sauvegarde le dataset final nettoyé dans `output/dataset_final_clean.csv`.

```python
from pathlib import Path
import pandas as pd

from scripts.data_pipeline import EnergyDataPipeline

PROJECT_ROOT = Path.cwd().resolve().parent if Path.cwd().name == "notebooks" else Path.cwd()
DATA_METEO = PROJECT_ROOT / "data" / "Data_Climat"
DATA_RTE = PROJECT_ROOT / "data" / "Data_eCO2"
OUTPUT_DIR = PROJECT_ROOT / "output"

print("Project root:", PROJECT_ROOT)
print("Meteo path:", DATA_METEO)
print("RTE path:", DATA_RTE)
print("Output path:", OUTPUT_DIR)
```

```python
pipeline = EnergyDataPipeline(
    meteo_path=DATA_METEO,
    rte_path=DATA_RTE,
    output_path=OUTPUT_DIR,
)

df_clean = pipeline.run(save_intermediate=False)

clean_file = OUTPUT_DIR / "dataset_final_clean.csv"
df_clean.to_csv(clean_file, index=False)

print("Saved:", clean_file)
print("Shape:", df_clean.shape)
df_clean.head()
```
