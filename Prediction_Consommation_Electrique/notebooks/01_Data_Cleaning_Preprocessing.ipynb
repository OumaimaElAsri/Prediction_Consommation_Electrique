## Étape 1: Nettoyage et Prétraitement des Données

Ce notebook contient les étapes initiales pour nettoyer et préparer les données.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sys
from pathlib import Path

# Ajouter le répertoire scripts au path
sys.path.insert(0, str(Path('.').resolve()))

from scripts.data_cleaning import load_and_clean_data
from scripts.data_preprocessing import preprocess_data
from scripts.utils import create_data_summary, save_data, get_project_paths

print("Imports réussis!")
```

### Charger les données

```python
# Remplacer 'votre_fichier.csv' par le chemin réel de vos données
# data_path = 'data/votre_fichier.csv'
# df = load_and_clean_data(data_path)

# Pour l'instant, créer un exemple de données
np.random.seed(42)
df = pd.DataFrame({
    'date': pd.date_range('2020-01-01', periods=100),
    'region': np.random.choice(['Île-de-France', 'Auvergne-Rhône-Alpes', 'Occitanie'], 100),
    'consommation_kwh': np.random.randint(1000, 5000, 100),
    'temperature': np.random.uniform(5, 30, 100),
    'jour_semaine': np.random.choice(['lundi', 'mardi', 'mercredi', 'jeudi', 'vendredi', 'samedi', 'dimanche'], 100)
})

print(f"Forme des données: {df.shape}")
print(f"\nPremières lignes:\n{df.head()}")
```

### Résumé des données

```python
summary = create_data_summary(df)
print("Résumé des données:")
for key, value in summary.items():
    if key != 'dtypes':
        print(f"{key}: {value}")
```

### Prétraitement des données

```python
# Prétraiter les données
df_processed = preprocess_data(df, scale=False)

print(f"Forme après prétraitement: {df_processed.shape}")
print(f"\nColonnes: {df_processed.columns.tolist()}")
print(f"\nAperçu:\n{df_processed.head()}")
```

### Sauvegarde des données traitées

```python
# Sauvegarder les données nettoyées et prétraitées
save_data(df_processed, 'data_cleaned.csv', directory='data')
print("Données sauvegardées!")
```
